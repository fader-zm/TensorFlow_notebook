```python
import random
```

```python
# 画出某城市11点到12点1小时内每分钟的温度变化折线图，温度范围在15度~18度

# 画出温度折线图

# 准备数据
x = range(60)
y_shanghai = [random.uniform(15, 18) for i in x]
# 创建画布
plt.figure(figsize=(20, 8), dpi=100)
# 添加自定义刻度
# 构造x刻度标签
x_ticks_label = [u"11点{}分".format(i) for i in x]
# 构造y轴刻度
y_ticks_lable = range(40)
plt.xticks(x[::5], x_ticks_label[::5])
plt.yticks(y_ticks_lable[::1])

# 添加网格显示
plt.grid(True, linestyle='--', alpha=0.5)

# 添加x轴/y轴/标题描述
plt.xlabel('时间')
plt.ylabel('温度')
plt.title("中午11点0分到12点之间的温度变化图示")
# 绘制图像
plt.plot(x, y_shanghai)
# 保存图像
plt.savefig('./image/shanghai_wendu_plot.png')
# 显示图像
plt.show()
```

```python
# 画出某城市11点到12点1小时内每分钟的温度变化折线图，温度范围在15度~18度

# 画出温度折线图

# 准备数据
x = range(60)
y_shanghai = [random.uniform(15, 18) for i in x]
y_beijing = [random.uniform(20, 30) for i in x]
# 创建画布
plt.figure(figsize=(20, 8), dpi=100)
# 添加自定义刻度
# 构造x刻度标签
x_ticks_label = [u"11点{}分".format(i) for i in x]
print(x_ticks_label)
# 构造y轴刻度
y_ticks_lable = range(40)
plt.xticks(x[::5], x_ticks_label[::5])
plt.yticks(y_ticks_lable[::1])

# 添加网格显示
plt.grid(True, linestyle='--', alpha=0.5)

# 添加x轴/y轴/标题描述
plt.xlabel('时间')
plt.ylabel('温度')
plt.title("中午11点0分到12点之间的温度变化图示")
# 绘制图像
plt.plot(x, y_shanghai)
plt.plot(x, y_beijing, color='r', linestyle='--', label='北京')
# 显示图例
plt.legend(loc='best')
# 保存图像
plt.savefig('./image/shanghai_wendu_plot.png')
# 显示图像
plt.show()
```

### 完整代码

```python
# 准备数据
x = range(60)
y_beijing = [random.uniform(20, 24) for i in x]
y_shanghai = [random.uniform(16, 19) for i in x]
# 创建画布
plt.figure(figsize=(20, 8), dpi=100)

# 自定义刻度 xticks, yticks
y_ticks_label = range(60)
x_ticks_label = [u"11点{}分".format(i) for i in x]
print(x_ticks_label)
plt.xticks(x[::5], x_ticks_label[::5])
plt.yticks(y_ticks_label[::1])
# x轴/y轴/标题显示xlabel, ylabel, title
plt.xlabel('时间')
plt.ylabel('温度')
plt.title('某城市11点-12点温度变化')
# 添加网格线 grid
plt.grid(linestyle='--', alpha=0.5)
# 显示图例 legend
plt.legend(loc='best')

# 绘制图像
plt.plot(x, y_beijing, color='r', linestyle='--', label='北京')
plt.plot(x, y_shanghai, color='b', linestyle='--', label='上海')

# 显示图像
plt.show()
```

### 多个坐标系显示-plt.subplots

plt.函数名()相当于面向过程的画图方法，axes.set_方法名()相当于面向对象的画图方法

```python
# 需求: 画出某城市11点到12点1小时内每分钟的温度变化折线图, 温度范围在15度到18度
```

```python
# 准备数据
x = range(60)
y_beijing = [random.uniform(15, 18) for i in x]
y_shanghai = [random.uniform(1, 14) for i in x]

# 创建画布
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(20, 8), dpi=100)
# 绘制图像
ax[0].plot(x, y_beijing, label='北京')
ax[1].plot(x, y_shanghai, label='上海')

# 添加x,y轴刻度
x_ticket_label = [u"11点{}分".format(i) for i in x]
y_tivket_label = range(40)

ax[0].set_xticks(x[::5])
ax[0].set_yticks(y_tivket_label[::5])
ax[0].set_xticklabels(x_ticket_label[::5])

ax[1].set_xticks(x[::5])
ax[1].set_yticks(y_tivket_label[::5])
ax[1].set_xticklabels(x_ticket_label[::5])

# 设置x, y轴标签和标题
ax[0].set_xlabel("时间")
ax[0].set_ylabel("温度")
ax[0].set_title("上海中午11点--12点温度变化图", fontsize=25)

ax[1].set_xlabel("时间")
ax[1].set_ylabel("温度")
ax[1].set_title("上海中午11点--12点温度变化图", fontsize=25)

# 显示图例
ax[0].legend(loc='best')
ax[1].legend(loc='best')

# 添加网格显示
ax[0].grid(True, linestyle='--', alpha=0.5)
ax[1].grid(True, linestyle='--', alpha=0.5)

# 显示图形
plt.show()
```
