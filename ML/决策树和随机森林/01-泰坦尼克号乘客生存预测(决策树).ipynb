```python
# 1.获取数据
data = pd.read_csv('http://biostat.mc.vanderbilt.edu/wiki/pub/Main/DataSets/titanic.txt')
```

```python
# 2.数据基本处理  特征选择，缺失值处理，划分训练集和测试集
# 2.1特征选择
x = data[['pclass', 'age', 'sex']]  # 获取特征值
y = data['survived'] # 获取目标值
```

```python
x.head()
```

```python
# 2.2 缺失值处理 填充缺失值
x['age'].fillna(x['age'].mean(), inplace=True)
```

```python
# 2.3 划分训练集和测试集
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
x_train.head()
```

```python
x_train.to_dict(orient='records')
```

```python
# 3.特征工程 one_hot编码 字典特征抽取的方式
transfer = DictVectorizer(sparse=False)

x_train = transfer.fit_transform(x_train.to_dict(orient='records'))  # [{},{}]

x_test = transfer.transform(x_test.to_dict(orient='records'))
```

```python
# 4.机器学习 决策树 
# 4.1 创建模型 初始化估计器
estimator = DecisionTreeClassifier(criterion='gini', max_depth=5)
```

```python
# 4.2 训练模型 fit
estimator.fit(x_train, y_train)
```

```python
# 5.模型评估 准确率
estimator.score(x_test, y_test)
```

```python
# 决策树可视化
# 参数：决策树对象，文件保存路径，特征名字列表
export_graphviz(estimator, out_file='./tree.dot', feature_names=transfer.get_feature_names())
```

# 随机森林实现

```python
# 1.获取数据
data = pd.read_csv('http://biostat.mc.vanderbilt.edu/wiki/pub/Main/DataSets/titanic.txt')
```

```python
# 2.数据基本处理  特征选择，缺失值处理，划分训练集和测试集
# 2.1特征选择
x = data[['pclass', 'age', 'sex']]  # 获取特征值
y = data['survived'] # 获取目标值
```

```python
# 2.2 缺失值处理 填充缺失值
x['age'].fillna(x['age'].mean(), inplace=True)
```

```python
# 2.3 划分训练集和测试集
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
x_train.head()
```

```python
# 3.特征工程 one_hot编码 字典特征抽取的方式
transfer = DictVectorizer(sparse=False)
```

```python
x_train = transfer.fit_transform(x_train.to_dict(orient='records'))  # [{},{}]
```

```python
x_test = transfer.transform(x_test.to_dict(orient='records'))
```

```python
# 4.机器学习 随机森林
# 4.1 创建模型 初始化估计器
# estimator = DecisionTreeClassifier(criterion='gini', max_depth=5)
# estimator_rf = RandomForestClassifier(n_estimators=100, max_depth=5)
estimator_rf = RandomForestClassifier()
```

```python
# 交叉验证+网格搜索
# 构建参数字典
param_dict = {'n_estimators':[100, 200, 500], 'max_depth':[3,5]}
estimator_gscv = GridSearchCV(estimator=estimator_rf, param_grid=param_dict, cv=4)
```

```python
# 4.2 训练模型 fit
estimator_gscv.fit(x_train, y_train)
```

```python
# 5.模型评估 准确率
estimator_gscv.score(x_test, y_test)
```

```python
# 获取最优的随机森林
estimator_gscv.best_estimator_
```
